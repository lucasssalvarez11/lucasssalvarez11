````python

BASE_DIR = Path("/content/drive/MyDrive/Colab Notebooks/Lab Programacion Cientifica /Proyecto /entrega3")

DATAREAL_DIR = BASE_DIR / "raw"
PRED_DIR = BASE_DIR / "predictions"

print("BASE_DIR     :", BASE_DIR)
print("DATAREAL_DIR :", DATAREAL_DIR)
print("PRED_DIR     :", PRED_DIR)
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
````

````python
clientes_path = DATAREAL_DIR / "clientes.parquet"
productos_path = DATAREAL_DIR / "productos.parquet"

clientes_df = pd.read_parquet(str(clientes_path))
productos_df = pd.read_parquet(str(productos_path))

universe_size = len(clientes_df) * len(productos_df)
print("Tama√±o universo (clientes x productos):", universe_size)


def evaluar_predicciones_universo(batch_id: int, predcsv_name: str):

    batch_path = DATAREAL_DIR / f"batch_t{batch_id}.parquet"
    print(f"\n=== Evaluando batch {batch_id} ===")
    print("Archivo real (batch):", batch_path)

    data = pd.read_parquet(str(batch_path))

    data_pairs = data.groupby(["customer_id", "product_id"]).size().reset_index()
    true_pairs = set(map(tuple, data_pairs[["customer_id", "product_id"]].values))

    pred_path = PRED_DIR / predcsv_name
    print("Archivo de predicciones:", pred_path)

    preds_raw = pd.read_csv(str(pred_path))

    preds_ids = preds_raw.iloc[:, :2].copy()
    preds_ids.columns = ["customer_id", "product_id"]

    preds_ids = preds_ids.drop_duplicates()
    pred_pairs = set(map(tuple, preds_ids[["customer_id", "product_id"]].values))

    print("Predicciones realizadas: ", len(pred_pairs))
    print("Universo (clientes x productos): ", universe_size)

    # TP = predicho y real
    tp = len(pred_pairs & true_pairs)
    # FP = predicho pero no real
    fp = len(pred_pairs - true_pairs)
    # FN = real pero no predicho
    fn = len(true_pairs - pred_pairs)
    # TN = universo total - todo lo dem√°s
    tn = universe_size - tp - fp - fn


    y_true = [1] * tp + [0] * fp + [1] * fn + [0] * tn
    y_pred = [1] * tp + [1] * fp + [0] * fn + [0] * tn

    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
    recall_val = recall_score(y_true, y_pred, zero_division=0)
    balanced_acc = (recall_val + specificity) / 2

    print("-" * 50)
    print("Compras efectivas (TP + FN):", tp + fn)
    if tp + fn > 0:
        print("Porcentaje de compras efectivas recomendadas:", tp / (tp + fn) * 100)
    else:
        print("No hay compras efectivas en este batch (TP+FN = 0).")

    metrics = {
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_val,
        "f1": f1_score(y_true, y_pred, zero_division=0),
        "specificity": specificity,
        "balanced_accuracy": balanced_acc,
        "TP": tp,
        "FP": fp,
        "FN": fn,
        "TN": tn,
    }

    print("RESULTADOS M√âTRICAS")
    print("==============================================")
    print(f"TP (Verdaderos Positivos): {tp}")
    print(f"FP (Falsos Positivos):     {fp}")
    print(f"FN (Falsos Negativos):     {fn}")
    print(f"TN (Verdaderos Negativos): {tn}")
    print("----------------------------------------------")
    print(f"Accuracy:          {metrics['accuracy']:.4f}")
    print(f"Precision:         {metrics['precision']:.4f}")
    print(f"Recall (TPR):      {metrics['recall']:.4f}")
    print(f"Specificity (TNR): {metrics['specificity']:.4f}")
    print(f"Balanced Accuracy: {metrics['balanced_accuracy']:.4f}")
    print(f"F1 Score:          {metrics['f1']:.4f}")

    # Matriz de confusi√≥n normalizada
    cm = pd.crosstab(
        pd.Series(y_true, name="Real"),
        pd.Series(y_pred, name="Pred"),
        normalize="index",
    )

    print("Matriz de Confusi√≥n (normalizada por fila):")
    print(cm, "\n")

    plt.figure(figsize=(6, 5))
    sns.heatmap(
        cm, annot=True, fmt=".2f",
        cmap="Blues",
        vmin=0, vmax=1
    )
    plt.title("Matriz de Confusi√≥n (Normalizada) - Universo cliente x producto")
    plt.xlabel("Predicci√≥n")
    plt.ylabel("Real")
    plt.show()

    return metrics
````

````python
def comparar_clientes(batch_id, predcsv_name):
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns


    pred_path = PRED_DIR / predcsv_name
    print(f"\n=== Comparando clientes para batch {batch_id} ===")
    print("Archivo de predicciones:", pred_path)

    preds_raw = pd.read_csv(str(pred_path))

    preds = preds_raw.iloc[:, :2].copy()
    preds.columns = ["customer_id", "product_id"]

    preds = preds.drop_duplicates()

    pred_counts = preds["customer_id"].value_counts().reset_index()
    pred_counts.columns = ["customer_id", "pred_count"]

    batch_path = DATAREAL_DIR / f"batch_t{batch_id}.parquet"
    print("Archivo real (batch):", batch_path)

    batch_data = pd.read_parquet(str(batch_path))

    true_counts = batch_data["customer_id"].value_counts().reset_index()
    true_counts.columns = ["customer_id", "true_count"]

    merged_counts = pred_counts.merge(true_counts, on="customer_id", how="outer").fillna(0)
    merged_counts["pred_count"] = merged_counts["pred_count"].astype(int)
    merged_counts["true_count"] = merged_counts["true_count"].astype(int)

    plt.figure(figsize=(12, 6))

    sns.histplot(
        merged_counts["true_count"],
        color="blue",
        label="Compras reales",
        alpha=0.5,
        bins=30,
    )

    sns.histplot(
        merged_counts["pred_count"],
        color="orange",
        label="Predicciones",
        alpha=0.5,
        bins=30,
    )

    plt.title(f"Histograma Comparativo: Compras Reales vs Predicciones (Batch {batch_id})")
    plt.xlabel("N√∫mero de productos por cliente")
    plt.ylabel("Cantidad de clientes")
    plt.legend()
    plt.tight_layout()
    plt.show()

    return merged_counts
````

Por lo tanto, un siguiente paso natural ser√≠a explorar ajustes de umbral de probabilidad y/o restricciones de tipo top-K por cliente para mejorar la precisi√≥n y, con ello, el F1, sin perder demasiada cobertura.

### üìå An√°lisis comparativo [2.0 puntos]

Agregamos funciones auxiliares para generar graficos comparativos

````python
def compare_results(res1, res2, res3, res4):
    """
    Compara las m√©tricas de 4 batches y genera tablas + gr√°ficos.

    """

    resultados = [res1, res2, res3, res4]
    df_res = pd.DataFrame(resultados)
    df_res["batch"] = ["Batch 1", "Batch 2", "Batch 3", "Batch 4"]

    cols_orden = [
        "batch",
        "accuracy",
        "precision",
        "recall",
        "specificity",
        "balanced_accuracy",
        "f1",
        "TP", "FP", "FN", "TN"
    ]
    cols_orden = [c for c in cols_orden if c in df_res.columns]
    df_res = df_res[cols_orden]

    print("#" * 10 + " M√©tricas comparativas entre batches " + "#" * 10)
    print(df_res)
    print("#" * 57)

    if "f1" in df_res.columns:
        plt.figure(figsize=(8, 4))
        sns.barplot(data=df_res, x="batch", y="f1", palette="Blues")
        plt.title("F1 Score por Batch")
        plt.ylim(0, 1)
        plt.tight_layout()
        plt.show()

    metricas_plot = [m for m in ["precision", "recall", "specificity"] if m in df_res.columns]

    if metricas_plot:
        df_melt = df_res.melt(
            id_vars="batch",
            value_vars=metricas_plot,
            var_name="m√©trica",
            value_name="valor"
        )

        plt.figure(figsize=(10, 4))
        sns.barplot(data=df_melt, x="batch", y="valor", hue="m√©trica", palette="Set2")
        plt.title("M√©tricas por Batch")
        plt.ylim(0, 1)
        plt.tight_layout()
        plt.show()

    return df_res
````

````python
df_comp_batches = compare_results(metrics_b1, metrics_b2, metrics_b3, metrics_b4)
df_comp_batches
````

````python
def compare_results_advanced(res1, res2, res3, res4, batch_labels=None):
    """
    Compara m√©tricas de 4 batches con foco en F1
    y genera varios gr√°ficos descriptivos con ejes Y ajustados
    al rango real de cada m√©trica.
    """
    resultados = [res1, res2, res3, res4]
    df_res = pd.DataFrame(resultados)

    if batch_labels is None:
        batch_labels = ["Batch 1", "Batch 2", "Batch 3", "Batch 4"]
    df_res["batch"] = batch_labels

    for col in ["TP", "FP", "FN", "TN"]:
        if col not in df_res.columns:
            df_res[col] = 0

    cols_orden = [
        "batch",
        "accuracy", "precision", "recall", "specificity",
        "balanced_accuracy", "f1",
        "TP", "FP", "FN", "TN"
    ]
    cols_orden = [c for c in cols_orden if c in df_res.columns]
    df_res = df_res[cols_orden]

    df_res["recomendaciones_totales"] = df_res["TP"] + df_res["FP"]
    df_res["compras_reales_totales"] = df_res["TP"] + df_res["FN"]

    df_res["fp_rate_recomendaciones"] = df_res.apply(
        lambda row: row["FP"] / row["recomendaciones_totales"]
        if row["recomendaciones_totales"] > 0 else 0,
        axis=1
    )

    df_res["fn_rate_reales"] = df_res.apply(
        lambda row: row["FN"] / row["compras_reales_totales"]
        if row["compras_reales_totales"] > 0 else 0,
        axis=1
    )

    print("#" * 10 + " M√©tricas comparativas entre batches (avanzado) " + "#" * 10)
    print(df_res)
    print("#" * 70)

    def get_ylim_for_metric(series):
        vals = series.dropna().astype(float)
        if len(vals) == 0:
            return (0, 1)
        min_v = vals.min()
        max_v = vals.max()
        if min_v == max_v:
            margin = 0.05
        else:
            margin = max(0.02, (max_v - min_v) * 0.2)
        y_min = max(0, min_v - margin)
        y_max = min(1, max_v + margin)
        if y_min == y_max:
            y_min = max(0, y_min - 0.1)
            y_max = min(1, y_max + 0.1)
        return (y_min, y_max)

    if "f1" in df_res.columns:
        plt.figure(figsize=(6, 4))
        sns.barplot(data=df_res, x="batch", y="f1", palette="Blues")
        plt.title("F1 Score por Batch")
        y_min, y_max = get_ylim_for_metric(df_res["f1"])
        plt.ylim(y_min, y_max)
        plt.tight_layout()
        plt.show()

    for metric_name, title in [
        ("precision", "Precision por Batch"),
        ("recall", "Recall por Batch"),
        ("specificity", "Specificity por Batch"),
    ]:
        if metric_name in df_res.columns:
            plt.figure(figsize=(6, 4))
            plt.plot(
                df_res["batch"],
                df_res[metric_name],
                marker="o",
                linestyle="-"
            )
            plt.title(title)
            y_min, y_max = get_ylim_for_metric(df_res[metric_name])
            plt.ylim(y_min, y_max)
            plt.xlabel("Batch")
            plt.ylabel(metric_name.capitalize())
            plt.tight_layout()
            plt.show()

    df_err = df_res[["batch", "fp_rate_recomendaciones", "fn_rate_reales"]].melt(
        id_vars="batch",
        var_name="tipo_error",
        value_name="tasa"
    )

    df_err["tipo_error"] = df_err["tipo_error"].map({
        "fp_rate_recomendaciones": "% FP sobre recomendaciones",
        "fn_rate_reales": "% FN sobre compras reales"
    })

    plt.figure(figsize=(8, 4))
    sns.barplot(data=df_err, x="batch", y="tasa", hue="tipo_error", palette="Set3")
    plt.title("Tasas de error por Batch")
    y_min, y_max = get_ylim_for_metric(df_err["tasa"])
    plt.ylim(y_min, y_max)
    plt.ylabel("Tasa")
    plt.xlabel("Batch")
    plt.legend(title="")
    plt.tight_layout()
    plt.show()

    return df_res
````

````python
df_res_avanzado = compare_results_advanced(
    metrics_b1,
    metrics_b2,
    metrics_b3,
    metrics_b4,
    batch_labels=["T-1", "T-2", "T-3", "T-4"]
)
df_res_avanzado
````
